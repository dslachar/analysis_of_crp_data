## Summary Statistics

```python
import pandas as pd
```

```python
indiv = pd.read_pickle('indivs.pkl')
indiv16 = pd.read_pickle('indivs16.pkl')
frames = [indiv,indiv16]
indivs_df = pd.concat(frames,ignore_index=True)
```

```python
# change the dtype of Amount from object to integer
indivs_df['Amount'] = indivs_df['Amount'].apply(lambda x: int(x))

# group the dataframe by Cycle (Year)
cycle_df = indivs_df.groupby('Cycle')
```

```python
indivs_df
```

```python
def group_years(i):
    '''Extracts total amount of contributions from each year. Returns the year, total amount, and
    number of contributions'''

    year = i
    amount = cycle_df.get_group(i).Amount.sum()
    contrib = len(cycle_df.get_group(i))

    return [[year,amount,contrib]]
```

```python
# goes through each of the years to create the total sum dataframe
total_df = pd.DataFrame(columns=['Year','Total_Amount','Total_Contributions'])
years = ['1992','1996','2000','2004','2008','2012','2016']
for i in years:
    total_df = total_df.append(pd.DataFrame(group_years(i),columns=['Year','Total_Amount','Total_Contributions']))

# change datatype of total_contrib and reindex dataframe
total_df.Total_Contributions = total_df.Total_Contributions.astype(int)
total_df = total_df.reset_index(drop=True)
total_df
```

```python
# created a win/loss data frame
indivs_df.groupby(['Cycle','Amount','Party','Results']).size()
win_df = indivs_df.loc[indivs_df['Results'] == 'W']
loss_df = indivs_df.loc[indivs_df['Results'] == 'L']
frame = [win_df,loss_df]
win_loss = pd.concat(frame)

# took out the Unknown party and lumped together third party and other
win_loss = win_loss.loc[win_loss.Party != 'U']
win_loss.replace({'Party': {'3': 'I/O', 'O': 'I/O'}}, inplace=True)
```

```python
# took out the Unknown and PAC party and lumped together third party and other
indivs_df = indivs_df.loc[indivs_df.Party != 'U']
indivs_df = indivs_df.loc[indivs_df.Party != 'P']
indivs_df.replace({'Party': {'3': 'I/O', 'O': 'I/O'}}, inplace=True)
```

```python
# created a win/loss data frame
indivs_df.groupby(['Cycle','Amount','Party','Results']).size()
win_df = indivs_df.loc[indivs_df['Results'] == 'W']
loss_df = indivs_df.loc[indivs_df['Results'] == 'L']
frame = [win_df,loss_df]
win_loss = pd.concat(frame)

# took out the Unknown party and lumped together third party and other
win_loss = win_loss.loc[win_loss.Party != 'U']
win_loss.replace({'Party': {'3': 'I/O', 'O': 'I/O'}}, inplace=True)
```

```python
# create a win and loss dataframe
w_df=win_loss.loc[win_loss['Results'].apply(lambda x: x=='W')]
l_df=win_loss.loc[win_loss['Results'].apply(lambda x: x=='L')]
```

```python
# winning stats
w_df.groupby('Cycle')['Amount'].max()
w_df.loc[w_df.Amount > 0].groupby('Cycle')['Amount'].min()
w_df.groupby('Cycle')['Amount'].sum()
w_df.groupby('Cycle')['Amount'].median()
w_df.groupby('Cycle')['Amount'].mean()
```
